### Método de Newton.

```julia
function raiz_polinomio(xo)
    list=zeros(Complex64,100); #crea una lista con 100 espacios. 
    x=xo;
    for i in 1:100 
        f= x^3-x^2-x-1
        df=3*x^2-2*x-1 
        x=x-f/df;
        list[i]=x; # este comando llena la lista creada con los valores obtenidos de iterar el método de Newton por medio del for. 
    end
    return list
end
```

```julia
using PyPlot

```

```julia

list1=raiz_polinomio(-0.1); #listas de valores para cuatro distintas condiciones iniciales (parentesís).
list2=raiz_polinomio(1.1);
list3=raiz_polinomio(1.5);
list4=raiz_polinomio(1.8);
```

```julia
scatter(1:100,list1);
plot(1:100,list1,label=("-0.1"));# Se usa para graficar la cantidad de iteraciones contra los valores obtenidos para las raíces. 
scatter(1:100,list2);
plot(1:100,list2,label=("1.1"));
scatter(1:100,list3);
plot(1:100,list3,label=("1.5"));
scatter(1:100,list4);
plot(1:100,list4,label=("1.8"));
```

```julia

plot(real(list1),label=("-0.1"))
plot(real(list2),label=("1.1"))
plot(real(list3),label=("1.5"))
plot(real(list4),label=("1.8"))
    legend()
 xlabel("Iteración")
    ylabel("Raíz")
    title(" Newton Raphson")
    grid("on")
show(); #notemos que los valores que más oscilan son los correspondientes a la condición inicial -0.1 (color azul),
```

### Método de Bisección.

```julia
function biseccion(a,b)
    list=zeros(Complex64,100)
    p = (a+b)/2  #se aplica el método de la bisección, donde a y b son los extremos del intervalo que acota a la raíz de f(x). 
    f(x) = x^3-x^2-x-1
    for i in 1:100   #el ciclo for toma cien iteraciones para llenar la lista creada. 
        if  f(a)*f(p) > 0 #si se cumple la deigualdad el valor p será a. 
            a = p           
        end
        
        if f(b)*f(p) > 0 #si pasa la desigualdad se cumplira que p es b. 
            b = p
        end
        p = (a+b)/2 #esto indica el nuevo valor de p, que genera un intervalo más pequeño que el original que acota la raíz buscada.
       list[i]=p 
    end
    return list
end;
```

Creamos cuatro listas con distintos intevalos para ver como convergen al valor de la raíz por medio de sus gráficas.

```julia
lista1=biseccion(-7,10); #listas de valores para cuatro intervalos distintos. 
lista2=biseccion(0,50);
lista3=biseccion(-100,100);
lista4=biseccion(1.3,1.9);
```

```julia
plot(real(lista1),label=("(-7,10)"))
plot(real(lista2),label=("(0,50)"))
plot(real(lista3),label=("(-100,100)")) #para este intevalo se nota que la gráfica tarda más en converger dada su lejania con la raíz.
plot(real(lista4),label=("(1.3,1.9)"))
    legend()
 xlabel("Iteración")
    ylabel("Raíz")
    title("Bisección")
    grid("on")
show();
```

```julia

function metodo_newton2(x0, f1::Function ,fprim1::Function)
list=(x=x0;f(x)=f1(x);fprim(x)=fprim1(x); metodo_newton.(x,f,fprim)) #construimos la función metodo_newton2 que hace uso de metodo_newton para crear una lista.
    t=[]
    push!(t,list[1])  #se genera una lista t y se introduce el primer valor de list en ella. 
    for i in 1:length(list) 
        k=0 #el ciclo for significa que el indice i y j de uno al último espacio de la lista "list" y "t" si se tiene que el valor absoluto de la diferencia de los valores es menor que epsilón se  sustituira el valor i de list en la lista de t hasta completarse. 
        for j in 1:length(t) 
       if abs(t[j]-list[i])<=0.000000001
                k=k+1
            end   
        end
        if k==length(t)
            push!(t,list[i])
        end
    end
return t
end
```

```julia
function metodo_newton3(y, f::Function) #se le pide al usuario solo la función y el valor inicial.
    x=symbols("x")   #se crea una variable simbólica y se sustituye en f pedida al usuario.
    f(x)
    fprim=diff(f(x),x)
    juliaprim=fjulia=lambdify(fprim,[x]); #como sabemos el comando lambdify cambierte una función de variables simbólicas a una de Julia y así se puede ya incluir en el ciclo de for para generar las iteraciones.
    for i in 1:5    
        y=y-f(y)/juliaprim(y)  
    end
    return y
end
```

```julia
(x=linspace(-.1,100,100);f(x)=x^2-2; metodo_newton3.(x,f))
```

```julia
function metodo_newton4(y0,f2::Function)
list2=(y=y0;f(x)=f2(x); metodo_newton3.(x,f))
    eps=1e^-8
    t=[]
    push!(t,list2[1])
    for i in 1:length(list2)  
        contador=0
        for j in 1:length(t)
            if abs(t[j]-list2[i])<=eps
                contador=contador+1
            end
        end
        if contador==length(t)
            push!(t,list2[i])
        end
    end
return t
end
```
